### EN ENTREPRISES EN AS BESOIN D'ENCAPSULER LE RESULTATS PRECEDENT AFIN DE L'AUTOMATISER ON PROCEDE AINSI 

```python
def extract_data_from_coingecko():
    url = "https://api.coingecko.com/api/v3/coins/markets"
    parameters = {
        "vs_currency": "usd",
        "order": "market_cap_desc",
        "per_page": 10000,
        "page": 1,
        "sparkline": False
    }
    response = requests.get(url, params=parameters)
    if response.status_code == 200:
        data = response.json()
        df = pd.json_normalize(data)
        return df
    else:
        raise Exception(f"Erreur lors de la requette de l'API coingecko: {response.status_code}")
```

```python
#resultat final de l'extraction des données de l'API coingecko
df_crypto = extract_data_from_coingecko() #afficher les informations sur le dataframe pour pouvoir les utiliser dans le projet ETL
df_crypto.info()
df_crypto.shape    #peremet de voir le nombre de lignes et de colonnes du dataframe pour pouvoir les utiliser dans le projet ETL
```

```python
#deuxieme transformation renommer les colonnes du dataframe 

df_transformed.columns = ["id", "symbol", "name", "price_usd", "market_cap_usd", "volume_24h_usd", "price_change_percentage_24h", "date"]
df_transformed
```

```python
#transformer la colone date pour qu 'elle soit au format datetime pour pouvoir l'utiliser par l'equipe de data analyste
df_transformed["date"] = pd.to_datetime(df_transformed["date"]).dt.date
df_transformed
```

```python
# la derniere transfo c'est de remplacer toutes les valeuur manquantes par des 0 pour que l'equipe de data analyste puisse utiliser les données sans avoir des erreurs de valeurs manquantes
df_transformed.fillna(0, inplace=True)
df_transformed
```

```python
# pour automatiser le processus d'extraction et de transformation des données de l'API coingecko 
# on va creer une fonction qui va faire tout le processus d'extraction et de transformation des données
#  de l'API coingecko pour pouvoir l'utiliser dans le projet ETL
def transform_data(df):
    df_transformed = df[["id", "symbol", "name", "current_price", "market_cap", "total_volume", "price_change_percentage_24h", "last_updated"]]
    df_transformed.columns = ["id", "symbol", "name", "price_usd", "market_cap_usd", "volume_24h_usd", "price_change_percentage_24h", "date"]
    df_transformed["date"] = pd.to_datetime(df_transformed["date"]).dt.date
    df_transformed.fillna(0, inplace=True)
    return df_transformed   
```

```python
#affichage de la base de donne transformee par automatisation

df_transformed_auto = transform_data(df_crypto)
df_transformed_auto
```

```python
df_transformed_auto.to_csv("data_transformed.csv", index=False)

#permet d'enregistrer le dataframe transformé dans un fichier csv 
```
